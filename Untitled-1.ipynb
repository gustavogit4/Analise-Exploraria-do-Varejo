```python
# Parâmetros
num_samples = 1000
sample_sizes = [2, 30, 1000]

# Crie uma função para gerar amostras de diferentes distribuições
def generate_samples(dist_name, sample_size):
    if dist_name == "binomial":
        return np.random.binomial(n = 5, p=0.2, size=sample_size)
    elif dist_name == "gamma":
        return np.random.gamma(shape=0.5,scale=2,size = sample_size)
    elif dist_name == "uniform":
        return np.random.uniform(1, 10, size=sample_size)

# Crie os gráficos
for dist_name in ["binomial", "gamma", "uniform"]:
    plt.figure(figsize=(12, 10))
    for i, sample_size in enumerate(sample_sizes):
        plt.subplot(2, 3, i + 1)
        samples = [generate_samples(dist_name, sample_size) for _ in range(num_samples)]
        sample_means = np.mean(samples, axis=1)
        
        # Trace o histograma das médias amostrais
        sns.histplot(sample_means, label='Sample Means')    
        
        plt.title(f'Sample Size: {sample_size}')
        plt.xlabel('Sample Mean')
        plt.ylabel('Frequency')
        plt.legend()

    plt.suptitle(f'{dist_name.capitalize()} Distribution - Sample Means with Normal Distribution', fontsize=16)
    plt.tight_layout()
    plt.subplots_adjust(top=0.9)

    plt.show()
```
